NumPy major contributions are:

- Efficient numeric computation with C primitives
- Efficient collections with vectorized operations


### Additional Note:  Usage of -1 in Numpy
---

In NumPy, the use of `-1` allows you to resize an array based on one dimension while letting NumPy automatically compute the size for that dimension. This is especially useful when you want to either access elements from the end or reshape an array with an unknown dimension.

### Accessing Elements Using -1

When using `-1` for array access, it refers to the last element along a particular dimension. This can be helpful to access elements near the end of an array.

```python
import numpy as np

# Example array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Access the last element along the first dimension
last_row = arr[-1, :]  # Equivalent to arr[1, :]

# Access the last element along the second dimension
last_column = arr[:, -1]  # Equivalent to arr[:, 2]
```

### Reshaping with -1

When using `-1` for reshaping, it allows NumPy to automatically compute the size of the dimension based on the other dimensions. This is especially useful when you want to reshape an array with an unknown dimension.

```python


# Example array
arr = np.array([1, 2, 3, 4, 5, 6])

# Reshape the array to a 2x3 array
reshaped_arr = arr.reshape(2, 3)

# Reshape the array with automatic computation of the second dimension
reshaped_auto_dim = arr.reshape(3, -1)  # NumPy computes the second dimension as 2
```


## 4. Numpy references are not Deep Copies 
---

When to take a slice of an numpy array or just assign the array to another variable, then the numpy arrays are not deep copied. 

#### Deep Copy

When the entire object alsong with its contents are copied to another location and can be identified by another variable

#### Shallow Copy

Only a copy of the object pointer is made and assigned to the new variable. The contents pointed by both the pointers will be the same 


image.png

[Image Reference](https://stackoverflow.com/questions/184710/what-is-the-difference-between-a-deep-copy-and-a-shallow-copy)

````python
from pprint import pprint
## Shallow Copy
a = np.ones((3,3))
pprint(a)
b = a 
## Now, if i change the value of b, then the value of a will also get affected
b[1,1] = 1000

print("Array B : ")
print(b)

print("Array A : ")
print(a)

````

To take an actual deep copy of an numpy array, use `copy()` method 

````python
a = np.ones((3,3))
pprint(a)
b = a.copy() ## deep copy
## Now, if i change the value of b, then the value of a will also get affected
b[1,1] = 1000

print("Array B : ")
print(b)

print("Array A : ")
print(a)
````

## 5. Mathematical Operations
---


We can directly perform operations such as addition, subraction, multiplication directly on the numpy array , rather than using traditional for loops

Note : The Dimensions of both the numpy arrays should match while performing mathematical operations

````python
import numpy as np
A = np.random.randint(1,10,(5,5))

print(A)
## Addition 
A = A + 5

print(A)
# Every Element Subraction
B = A - 5

# every element Multiplication
C = A * 5 

# Every element Division
D = A/2


## Every Element exponentiation
E = A**3

## Every element trignometric Operations
F = np.cos(A)

````

#### Matrix Multiplication 



````python
A = np.random.rand(3,4)
B = np.random.rand(4,3)

## Use function
C = np.matmul(A,B)

## use the symbol
C = A@B
````

#### Linear Algebra Functions 
---

Some of the linear algebra functions that are available within numpy are 

* determinant - `determinant()`
* Eigen Values and Vector - `eigs()`
* Singular Value Decomposition - `svd()`
* Norm of a Vector or Matrix - `norm()`

````python
A = np.random.rand(4,4)

print("Determinant of A " , np.linalg.det(A))

E,W = np.linalg.eig(A)   ## Eigen Value and Eigen Vector ( normalised )


## SVD
U,S,V = np.linalg.svd(A)


## Norm of vector 
a = np.array([1,2,3,4,5])

print("Norm of vector A : " , np.linalg.norm(a,2))

## Norm of Matrix
print("Norm of Matrix A : " , np.linalg.norm(A,2))
````

## 6. Statistics
---


````python
## Minimum of Array 
a = np.random.randint(0,20,20)
print(" The Array is : ", a)

## Minimum
minVal = np.min(a)
print("minimum Value of Array : " , np.min(a))

## maximum Val
maxVal = np.max(a)
print("maximum Value of Array : " , np.max(a))

## Sum 
sum = np.sum(a)
print("maximum Value of Array : " , np.sum(a))


## Mean
mean = np.mean(a)
print("mean Value of Array    : " , np.mean(a))

## stddev
stddev = np.std(a)
print("Std Dev of Array       : " , np.std(a))
````
